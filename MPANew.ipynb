Splitting into training and test sets

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=26)
```

Handling imbalanced Data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ovsamp = RandomOverSampler(random_state=59, sampling_strategy='not majority')

x_res, y_res = ovsamp.fit_resample(xtrain,ytrain)
```

Dealing with categorical data

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwordlist = stopwords.words('english')
stopwordlist.remove('not')
stopwordlist.remove("won")
print(stopwordlist)
```

```python
from nltk.stem.porter import PorterStemmer
corpus = []
titcorpus = []
dircorp = []
countrycorp = []
listedcorp = []

for i in range(0,len(newdf['rating'])):
    descrip = re.sub('[^a-zA-Z]', ' ', newdf['description'][i])
    title = newdf['title'][i].lower().split()
    director = newdf['director'][i].lower()
    country = newdf['country'][i].lower().split(",")
    listed = newdf['listed_in'][i].lower().split(",")
    descrip = descrip.lower()
    descrip = descrip.split()
    ps = PorterStemmer()
    # Stemming the words
    descrip = [ps.stem(word) for word in descrip if not word in set(stopwordlist)]
    title = [ps.stem(word) for word in title if not word in set(stopwordlist)]
    director = [ps.stem(name) for name in director]
    country = [ps.stem(name) for name in country]
    listed = [ps.stem(name) for name in listed]
    descrip = ' '.join(descrip)
    country = ' '.join(country)
    #print(descrip)
    corpus.append(descrip)
    titcorpus.append(title)
    dircorp.append(director)
    countrycorp.append(country)
    listedcorp.append(listed)
    #print(countrycorp)
```

```python
# Creating a Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.compose import make_column_transformer

vectorizer = CountVectorizer()
transformer = make_column_transformer((vectorizer,'title'),(vectorizer,'director'),
                                      (vectorizer,'country'),(vectorizer,'listed_in'),
                                      (vectorizer,'description'))
```

```python
x = transformer.fit_transform(x_res)
y = vectorizer.fit_transform(y_res)
```
